```python
import tensorflow as tf
from tensorflow import keras

DOWNLOAD_ROOT = "http://download.tensorflow.org/data/"
FILENAME = "quickdraw_tutorial_dataset_v1.tar.gz"
filepath = keras.utils.get_file(FILENAME,
                                DOWNLOAD_ROOT + FILENAME,
                                cache_subdir="datasets/quickdraw",
                                extract=True)
```

```python
file_path = '/root/.keras/datasets/quickdraw/eval.tfrecord-00001-of-00010'

for record in tf.data.TFRecordDataset([file_path]):
    example = tf.train.Example()
    example.ParseFromString(record.numpy())
    print(example)
    break
```

```python
def parse_record(records):
    feature_description = {
        'class_index': tf.io.FixedLenFeature([1], tf.int64),
        'ink': tf.io.VarLenFeature(dtype=tf.float32),
        'shape': tf.io.FixedLenFeature([2], tf.int64)
    }
    examples = tf.io.parse_example(records, feature_description)
    flat_sketches = tf.sparse.to_dense(examples['ink'])
    sketches = tf.reshape(flat_sketches, shape=[tf.size(records), -1, 3])
    lengths = examples['shape'][:, 0]
    labels = examples['class_index'][:, 0]
    return sketches, lengths, labels
```

```python
def dataset(filepaths, batch_size = 32):
    dataset = tf.data.TFRecordDataset(filepaths)
    dataset = dataset.shuffle(10000)
    dataset = dataset.batch(batch_size)
    dataset = dataset.map(parse_record)
    return dataset.prefetch(1)
train_dataset = dataset(train_files)
test_dataset = dataset(eval_files[:5])
valid_dataset = dataset(eval_files[5:])
```

```python
train_dataset
```

```python
for sketches , lengths , labels in train_dataset.take(1):
    print("sketches =", sketches)
    print("lengths =", lengths)
    print("labels =", labels)
```

```python
def crop_long_sketches(dataset, max_length=100):
    return dataset.map(lambda inks, lengths, labels: (inks[:, :max_length], labels))

cropped_train_set = crop_long_sketches(train_dataset)
cropped_valid_set = crop_long_sketches(valid_dataset)
cropped_test_set = crop_long_sketches(test_dataset)
```

```python
model = keras.Sequential([
    keras.layers.Conv1D(32, kernel_size=5, strides=2, activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.Conv1D(64, kernel_size=5, strides=2, activation='relu'),
    keras.layers.Conv1D(64, kernel_size=3, strides=2, activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.Conv1D(128, kernel_size=3, strides=2, activation='relu'),
    keras.layers.GRU(128, return_sequences=True),
    keras.layers.GRU(128),
    keras.layers.Dense(len(classes), activation='softmax')
])

optimizer = keras.optimizers.SGD(lr=1e-2, clipnorm=1.)
model.compile(optimizer=optimizer, loss=keras.losses.SparseCategoricalCrossentropy(), metrics=["accuracy", "sparse_top_k_categorical_accuracy"])
model.fit(cropped_train_set, validation_data=cropped_valid_set, epochs=2)
```

```python
y_test = np.concatenate([labels for _, _, labels in test_dataset])
y_probas = model.predict(test_dataset)
```

```python
np.mean(keras.metrics.sparse_top_k_categorical_accuracy(y_test, y_probas))
```
